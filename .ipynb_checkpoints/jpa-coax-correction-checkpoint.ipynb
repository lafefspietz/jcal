```python
datafolder = "jpa-nopump-scan2-s1p-coax-corrected"
```

```python
cyclefactor = np.exp(2*np.pi*1j*cycles_per_ghz*fghz)
```

```python
for index in range(len(data_corrected)):
    data_coax_corrected_trace =  data_corrected['trace' + str(index)].s[:,0,0]*cyclefactor
    data_coax_corrected_trace_network = rf.Network(frequency=fghz*1e9, s=data_coax_corrected_trace, z0=50)
    data_coax_corrected_trace_network.write_touchstone(datafolder + '/trace' + str(index) + '.s1p')
```
