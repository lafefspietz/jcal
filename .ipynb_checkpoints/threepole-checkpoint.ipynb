```python
import numpy as np
import matplotlib.pyplot as plt
import skrf as rf
from skrf.media import *
from skrf import Frequency
import json
import time
```

```python
file_object = open("threepole-components.json", "r")
try:
    components = json.loads(file_object.read())
except ValueError:
    pass

fghz  = np.linspace(4,9,1000)
f = fghz*1e9
sopen = fghz*0j + 1
open_network = rf.Network(frequency=f, s=sopen, z0=50)
omega = 2*np.pi*f
YC1 = 1j*omega*components['C1']
YC2 = 1j*omega*components['C2']
YC3 = 1j*omega*components['C3']
ZL1 = 1j*omega*components['L1']
ZL2 = 1j*omega*components['L2']
ZL3 = 1j*omega*components['L3']

C1_network = rf.Circuit.ShuntAdmittance(f,YC1, name='YC1')
L1_network = rf.Circuit.SeriesImpedance(f,ZL1, name='ZL1')
C2_network = rf.Circuit.ShuntAdmittance(f,YC2, name='YC2')
L2_network = rf.Circuit.SeriesImpedance(f,ZL2, name='ZL2')
C3_network = rf.Circuit.ShuntAdmittance(f,YC3, name='YC3')
L3_network = rf.Circuit.SeriesImpedance(f,ZL3, name='ZL3')

threepole_network = L3_network**C3_network**L2_network**C2_network**L1_network**C1_network**open_network

trace = {}
trace['fghz'] = fghz.tolist()
trace['phase'] = threepole_network.s_deg[:,0,0].tolist()
file_object = open("threepole-trace.json", "w")
json_string = file_object.write(json.dumps(trace))
file_object.close()
plt.plot(fghz,threepole_network.s_deg[:,0,0])
plt.grid()
```

```python
while True:
    file_object = open("threepole-components.json", "r")
    try:
        components = json.loads(file_object.read())
    except ValueError:
        pass

    fghz  = np.linspace(4,9,400)
    f = fghz*1e9
    sopen = fghz*0j + 1
    open_network = rf.Network(frequency=f, s=sopen, z0=50)
    omega = 2*np.pi*f
    YC1 = 1j*omega*components['C1']
    YC2 = 1j*omega*components['C2']
    YC3 = 1j*omega*components['C3']
    ZL1 = 1j*omega*components['L1']
    ZL2 = 1j*omega*components['L2']
    ZL3 = 1j*omega*components['L3']

    C1_network = rf.Circuit.ShuntAdmittance(f,YC1, name='YC1')
    L1_network = rf.Circuit.SeriesImpedance(f,ZL1, name='ZL1')
    C2_network = rf.Circuit.ShuntAdmittance(f,YC2, name='YC2')
    L2_network = rf.Circuit.SeriesImpedance(f,ZL2, name='ZL2')
    C3_network = rf.Circuit.ShuntAdmittance(f,YC3, name='YC3')
    L3_network = rf.Circuit.SeriesImpedance(f,ZL3, name='ZL3')

    threepole_network = L3_network**C3_network**L2_network**C2_network**L1_network**C1_network**open_network

    trace = {}
    trace['fghz'] = fghz.tolist()
    trace['phase'] = threepole_network.s_deg[:,0,0].tolist()
    file_object = open("threepole-trace.json", "w")
    json_string = file_object.write(json.dumps(trace))
    file_object.close()
    time.sleep(0.1)
```

```python
file_object = open("threepole-components.json", "r")
try:
    components = json.loads(file_object.read())
except ValueError:
    pass

fghz  = np.linspace(4,9,1000)
f = fghz*1e9
sopen = fghz*0j + 1
open_network = rf.Network(frequency=f, s=sopen, z0=50)
omega = 2*np.pi*f
YC1 = 1j*omega*components['C1']*3.1
YC2 = 1j*omega*components['C2']*2.8
YC3 = 1j*omega*components['C3']*1.3
ZL1 = 1j*omega*components['L1']*3.4
ZL2 = 1j*omega*components['L2']*2.2
ZL3 = 1j*omega*components['L3']*3.6

C1_network = rf.Circuit.ShuntAdmittance(f,YC1, name='YC1')
L1_network = rf.Circuit.SeriesImpedance(f,ZL1, name='ZL1')
C2_network = rf.Circuit.ShuntAdmittance(f,YC2, name='YC2')
L2_network = rf.Circuit.SeriesImpedance(f,ZL2, name='ZL2')
C3_network = rf.Circuit.ShuntAdmittance(f,YC3, name='YC3')
L3_network = rf.Circuit.SeriesImpedance(f,ZL3, name='ZL3')

threepole_network = L3_network**C3_network**L2_network**C2_network**L1_network**C1_network**open_network

plt.plot(fghz,threepole_network.s_deg_unwrap[:,0,0])
plt.grid()
```
