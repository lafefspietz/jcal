```python
# load needed python libraries matplotlib, numpy, and scikit-rf
import matplotlib.pyplot as plt
import numpy as np
import skrf as rf
import json
```

```python
rawdata = rf.read_all('jpa-nopump-scan2-s1p')
```

```python
raw_db_s11 = []
for index in range(len(rawdata)):
    raw_db_s11.append(list(rawdata['trace' + str(index)].s_db[:,0,0]))
```

```python
file_object = open("raw-db-s11.js", "w")
file_object.write("rawdbdata=" + json.dumps(raw_db_s11) + ";")
file_object.close()
```

```python
raw_phase_s11 = []
for index in range(len(rawdata)):
    raw_phase_s11.append(list(rawdata['trace' + str(index)].s_deg[:,0,0]))
```
